```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points= np.arange(-5,5,0.01)
```

```python
dx,dy=np.meshgrid(points,points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx)+(np.sin(dy)))
```

```python
plt.imshow(z)
```

```python
plt.imshow(z) # shows the values in image format in rectangluar like matlab
plt.colorbar()
plt.title('Plot for sin(x)+sin(y)')
```

```python
# creating two arrays- numeric, one boolean array and checking the conditions to get the answer
A=np.array([1,2,3,4])
B=np.array([100,200,300,400])
condition=np.array([False,True,False,True])
```

```python
#Here a_val will be printed if condition is true else B_val will be printed

answer= [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)]
```

```python
answer
```

```python
#instead of the above code we can make use of WHERE - here which states where condition is met then choose A or else B
answer2=np.where(condition,A,B)
answer2
```

Array INPUT & OUTPUT 

```python
n=np.arange(5)
n
```

```python
m=np.arange(10)
m
```

```python
#saving n - a single file into myarray through command
np.save('myarray',n)
```

```python
#open and load the saved file
np.load('myarray.npy')
```

```python
#saving multiple files, i.e zipping the files np.savez and save the filename with .npz extension
#it saves the file on the disk
np.savez('M_files.npz',x=m,y=n)
```

```python
#np.load - loads the file in the variable
archieve=np.load('M_files.npz')
```

```python
archieve['x']
```

```python
archieve['y']
```

```python
# saving as a text file and loading the same
ar= np.array([[1,2,3],[4,5,6]])
ar
```

```python
np.savetxt('mytextarray.txt',ar)
```

```python
file=np.loadtxt('mytextarray.txt')
file
```

```python
# Another way by adding delimiter
np.savetxt('mytextarray.txt',ar,delimiter=',')
np.loadtxt('mytextarray.txt',delimiter=',')
```
